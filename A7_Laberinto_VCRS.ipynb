```python
# DIBUJO DE UN LABERINTO SOBRE UNA IMAGEN Y ALGORITMO DE RESOLUCIÓN

# IMPORTS PARA EL CÓDIGO
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as img

# VARIABLES
X = 180;
Y = 180;
A = 0;
B = 0;

# CREACIÓN DE LAS DOS IMAGENES DEL LABERINTO
imagen_bgr = np.zeros((Y, X, 1), dtype=np.uint8)
imagen_lab = np.zeros((Y, X, 1), dtype=np.uint8)

# VARIABLES DE COLORES
Blanco = 255;
Negro = 0;
Salida = 200;
Paso = 100;
Ubicacion = 124;

# RELLENO DE LA IMAGEN DEL LABERINTO PARA RESOLVER
imagen_bgr[20:40, 0:120] = (Blanco)
imagen_bgr[100:120, 20:140] = (Blanco)
imagen_bgr[20:160, 140:160] = (Blanco)
imagen_bgr[60:160, 20:40] = (Blanco)
imagen_bgr[140:160, 0:20] = (Blanco)
imagen_bgr[60:160, 100:120] = (Blanco)
imagen_bgr[40:60, 60:80] = (Blanco)
imagen_bgr[60:80, 20:80] = (Blanco)
imagen_bgr[140:160, 60:120] = (Blanco)
imagen_bgr[20:40, 0:20] = (Salida)
imagen_bgr[140:160, 0:20] = (Salida)
# RELLENO DE LA IMAGEN DEL LABERINTO ORIGINAL
imagen_lab[20:40, 0:120] = (Blanco)
imagen_lab[100:120, 20:140] = (Blanco)
imagen_lab[20:160, 140:160] = (Blanco)
imagen_lab[60:160, 20:40] = (Blanco)
imagen_lab[140:160, 0:20] = (Blanco)
imagen_lab[60:160, 100:120] = (Blanco)
imagen_lab[40:60, 60:80] = (Blanco)
imagen_lab[60:80, 20:80] = (Blanco)
imagen_lab[140:160, 60:120] = (Blanco)
imagen_lab[20:40, 0:20] = (Salida)
imagen_lab[140:160, 0:20] = (Salida)

# UBICACIÓN INICIAL EN EL LABERINTO
y1 = 20
y2 = 40
x1 = 140
x2 = 160
imagen_bgr[y1:y2,x1:x2] = (Ubicacion)
imagen_lab[y1:y2,x1:x2] = (Ubicacion)

# VERIFICACIÓN INICIAL DE POSICIÓN
# VERIFICACIÓN IZQUIERDA
if(imagen_bgr[y1,(x1-1)] == (Negro)):
    print("SI1");
# VERIFICACIÓN SUPERIOR
if(imagen_bgr[(y1-1),x1] == (Negro)):
    print("SI2");
# VERIFICACIÓN INFERIOR
if(imagen_bgr[y2,(x2-1)] == (Blanco)):
    print("SI3");
# VERIFICACIÓN DERECHA
if(imagen_bgr[(y2-1),x2] == (Negro)):
    print("SI4");

# CICLO WHILE REPETITIVO HASTA COMPLETAR
while(A == 0):
    B = 0;
    
    # VERIFICACION DE SALIDA (FIN DEL LABERINTO)
    # VERIFICACIÓN DERECHA
    if((imagen_bgr[(y2-1),x2] == (Salida)) & (B == 0)):
        print("TERMINADO_DER");
        A = 1;
        B = 1;
        x1 = x1+20;
        x2 = x2+20;
        imagen_bgr[y1:y2,x1:x2] = (Negro);
        imagen_bgr[y1:(y2-10),x1:(x2-10)] = (Blanco);
        imagen_bgr[(y1+10):y2,(x1+10):x2] = (Blanco);
    # VERIFICACIÓN SUPERIOR
    elif((imagen_bgr[(y1-1),x1] == (Salida)) & (B == 0)):
        print("TERMINADO_SUP");
        A = 1;
        B = 1;
        y1 = y1-20;
        y2 = y2-20;
        imagen_bgr[y1:y2,x1:x2] = (Negro);
        imagen_bgr[y1:(y2-10),x1:(x2-10)] = (Blanco);
        imagen_bgr[(y1+10):y2,(x1+10):x2] = (Blanco);
    # VERIFICACIÓN INFERIOR
    elif((imagen_bgr[y2,(x2-1)] == (Salida)) & (B == 0)):
        print("TERMINADO_INF");
        A = 1;
        B = 1;
        y1 = y1+20;
        y2 = y2+20;
        imagen_bgr[y1:y2,x1:x2] = (Negro);
        imagen_bgr[y1:(y2-10),x1:(x2-10)] = (Blanco);
        imagen_bgr[(y1+10):y2,(x1+10):x2] = (Blanco);
    # VERIFICACIÓN IZQUIERDA
    if((imagen_bgr[y1,(x1-1)] == (Salida)) & (B == 0)):
        print("TERMINADO_IZQ");
        A = 1;
        B = 1;
        x1 = x1-20;
        x2 = x2-20;
        imagen_bgr[y1:y2,x1:x2] = (Negro);
        imagen_bgr[y1:(y2-10),x1:(x2-10)] = (Blanco);
        imagen_bgr[(y1+10):y2,(x1+10):x2] = (Blanco);
    
    # VERIFICACION DE COLOR BLANCO (CAMINO LIBRE)
    # VERIFICACIÓN DERECHA
    if((imagen_bgr[(y2-1),x2] == (Blanco)) & (B == 0)):
        print("BlancoDER");
        x1 = x1+20;
        x2 = x2+20;
        imagen_bgr[y1:y2,x1:x2] = (Paso);
        B = 1;
    # VERIFICACIÓN SUPERIOR
    elif((imagen_bgr[(y1-1),x1] == (Blanco)) & (B == 0)):
        print("BlancoSUP");
        y1 = y1-20;
        y2 = y2-20;
        imagen_bgr[y1:y2,x1:x2] = (Paso);
        B = 1;
    # VERIFICACIÓN INFERIOR
    elif((imagen_bgr[y2,(x2-1)] == (Blanco)) & (B == 0)):
        print("BlancoINF");
        y1 = y1+20;
        y2 = y2+20;
        imagen_bgr[y1:y2,x1:x2] = (Paso);
        B = 1;
    # VERIFICACIÓN IZQUIERDA
    elif((imagen_bgr[y1,(x1-1)] == (Blanco)) & (B == 0)):
        print("BlancoIZQ");
        x1 = x1-20;
        x2 = x2-20;
        imagen_bgr[y1:y2,x1:x2] = (Paso);
        B = 1;   
    
    # VERIFICACION DE COLOR GRIS (CAMINO YA RECORRIDO)
    # VERIFICACIÓN DERECHA
    if((imagen_bgr[(y2-1),x2] == (Paso)) & (B == 0)):
        print("PasoDER");
        imagen_bgr[y1:y2,x1:x2] = (Negro);
        x1 = x1+20;
        x2 = x2+20;
        imagen_bgr[y1:y2,x1:x2] = (Paso);
        B = 1;
    # VERIFICACIÓN SUPERIOR
    elif((imagen_bgr[(y1-1),x1] == (Paso)) & (B == 0)):
        print("PasoSUP");
        imagen_bgr[y1:y2,x1:x2] = (Negro);
        y1 = y1-20;
        y2 = y2-20;
        imagen_bgr[y1:y2,x1:x2] = (Paso);
        B = 1;
    # VERIFICACIÓN INFERIOR
    elif((imagen_bgr[y2,(x2-1)] == (Paso)) & (B == 0)):
        print("PasoINF");
        imagen_bgr[y1:y2,x1:x2] = (Negro);
        y1 = y1+20;
        y2 = y2+20;
        imagen_bgr[y1:y2,x1:x2] = (Paso);
        B = 1;
    # VERIFICACIÓN IZQUIERDA
    elif((imagen_bgr[y1,(x1-1)] == (Paso)) & (B == 0)):
        print("PasoIZQ");
        imagen_bgr[y1:y2,x1:x2] = (Negro);
        x1 = x1-20;
        x2 = x2-20;
        imagen_bgr[y1:y2,x1:x2] = (Paso);
        B = 1;   
#FIN DEL CICLO

# CREACIÓN DE VENTANAS PARA IMAGENES
cv.imshow('Laberinto resuelto', imagen_bgr)
cv.imshow('Laberinto original', imagen_lab)
cv.waitKey(0)

# CIERRE DE VENTANAS
cv.destroyAllWindows()
```
